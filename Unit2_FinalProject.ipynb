```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
csvFile = open("/content/drive/MyDrive/Colab Notebooks/Unit 2 - Climate/CloudFraction1_2002 (3).csv", 'r')
csvReader = csv.reader(csvFile, delimiter=',')
data_lists = list(csvReader)
```

```python
print(data_lists[0])
```

```python
len(data_lists[0])
```

```python
len(data_lists)
```

```python
height = len(data_lists)
width = len(data_lists[0])
```

```python
grid = np.empty([height, width, 3], dtype=np.uint8) 
```

```python
all_values = [] # We will fill in this empty list with all the data values as we go

for r, row in enumerate(data_lists):
    for c, value in enumerate(row):
        if float(value) < 99999:
            all_values.append(float(value)) # This is where we fill in the list with all data values
        if float(value) <= 0.0:
            grid[r, c] = [90, 136, 209]
        elif float(value) <= 0.05:
            grid[r, c] = [94, 145, 224]
        elif float(value) <= 0.1:
            grid[r, c] = [99, 150, 230]
        elif float(value) <= 0.15:
            grid[r, c] = [104, 156, 237]
        elif float(value) <= 0.20:
            grid[r, c] = [109, 161, 242]
        elif float(value) <= 0.25:
            grid[r, c] = [115, 166, 245]
        elif float(value) <= 0.30:
            grid[r, c] = [122, 172, 250]
        elif float(value) <= 0.35:
            grid[r, c] = [128, 177, 255]
        elif float(value) <= 0.40:
            grid[r, c] = [135, 181, 255]
        elif float(value) <= 0.45:
            grid[r, c] = [140, 182, 250]
        elif float(value) <= 0.50:
            grid[r, c] = [148, 189, 255]
        elif float(value) <= 0.55:
            grid[r, c] = [156, 194, 255]
        elif float(value) <= 0.60:
            grid[r, c] = [163, 198, 255]
        elif float(value) <= 0.65:
            grid[r, c] = [179, 208, 255]
        elif float(value) <= 0.70:
            grid[r, c] = [187, 208, 242]
        elif float(value) <= 0.75:
            grid[r, c] = [196, 215, 245]
        elif float(value) <= 0.80:
            grid[r, c] = [201, 218, 245]
        elif float(value) <= 0.85:
            grid[r, c] = [211, 224, 245]
        elif float(value) <= 0.90:
            grid[r, c] = [230, 234, 240]
        elif float(value) <= 0.95:
            grid[r, c] = [230, 234, 240]
        elif float(value) <= 1.00:
            grid[r, c] = [255, 255, 255]
        else:
            grid[r, c] = [0, 0, 0]
```

```python
print(min(all_values))
print(max(all_values))
```

```python
plt.title("Cloud Fraction 2002")
plt.imshow(grid)
plt.axis('off')
plt.savefig("CloudFraction_2002.png",dpi = 1000)
plt.show()
```
